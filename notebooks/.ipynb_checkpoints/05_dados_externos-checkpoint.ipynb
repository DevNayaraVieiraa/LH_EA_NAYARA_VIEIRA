```python
# INTEGRA√á√ÉO DE DADOS EXTERNOS - BANVIC
# M√≥dulo para buscar e integrar dados p√∫blicos relevantes
# Desenvolvido para enriquecer an√°lises banc√°rias

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta
import requests
import json
import warnings
warnings.filterwarnings('ignore')

print("M√ìDULO DE DADOS EXTERNOS - BANVIC")
print("="*50)

# ============================================================================
# SIMULA√á√ÉO DE DADOS EXTERNOS
# ============================================================================

def buscar_dados_externos():
    """
    Busca dados p√∫blicos relevantes
    TODO: Implementar conex√£o com APIs reais quando dispon√≠vel
    """
    print("üåê INICIANDO AN√ÅLISE DE DADOS EXTERNOS")
    print("\nBuscando dados p√∫blicos relevantes...")
    
    # Simula√ß√£o da taxa USD/BRL
    print("‚Ä¢ Coletando cota√ß√µes USD/BRL...")
    datas = pd.date_range('2024-01-01', '2024-12-31', freq='D')
    usd_brl = pd.DataFrame({
        'data': datas,
        'cotacao_usd': np.random.normal(5.2, 0.3, len(datas)),
        'variacao_pct': np.random.normal(0, 2, len(datas))
    })
    usd_brl['cotacao_usd'] = np.abs(usd_brl['cotacao_usd'])
    usd_brl = usd_brl.reset_index(drop=True)
    
    # Simula√ß√£o da SELIC
    print("‚Ä¢ Coletando taxa SELIC...")
    selic_exemplo = pd.DataFrame({
        'data': pd.date_range('2024-01-01', '2024-12-31', freq='M'),
        'taxa_selic': np.random.uniform(10.5, 13.5, 12)
    })
    
    # Simula√ß√£o do IPCA
    print("‚Ä¢ Coletando dados do IPCA...")
    ipca_exemplo = pd.DataFrame({
        'data': pd.date_range('2024-01-01', '2024-12-31', freq='M'),
        'ipca_mensal': np.random.uniform(0.2, 0.8, 12),
        'ipca_acumulado': np.cumsum(np.random.uniform(0.2, 0.8, 12))
    })
    
    print("Dados externos coletados")
    
    return {
        'usd_brl': usd_brl,
        'selic': selic_exemplo,
        'ipca': ipca_exemplo
    }

def buscar_dados_selic():
    """
    Busca taxa SELIC do Banco Central
    Implementa√ß√£o para demonstra√ß√£o - usar dados simulados
    """
    try:
        # Em produ√ß√£o, conectaria com API do BCB
        # url = 'https://api.bcb.gov.br/dados/serie/bcdata.sgs.11/dados'
        
        # Para desenvolvimento, usamos dados simulados
        data_exemplo = {
            'data': '01/01/2024',
            'start': '2024-01-01',
            'end': '2024-12-31'
        }
        
        selic_exemplo = pd.DataFrame({
            'data': pd.date_range(data_exemplo['start'], data_exemplo['end'], freq='D'),
            'taxa_selic': np.random.uniform(10.75, 11.25, 
                len(pd.date_range(data_exemplo['start'], data_exemplo['end'], freq='D')))
        })
        
        # Criar diret√≥rio se necess√°rio
        import os
        if not os.path.exists('./dados/externos'):
            os.makedirs('./dados/externos')
        
        selic_exemplo.to_csv('./dados/externos/selic_url.csv', index=False)
        
        print("Dados SELIC coletados")
        return selic_exemplo
    
    except Exception as e:
        print(f"Erro ao buscar SELIC: {e}")
        return None

# ============================================================================
# AN√ÅLISE DE CORRELA√á√ÉO COM DADOS BANC√ÅRIOS
# ============================================================================

def gerar_dados_bancarios_sample():
    """Gera sample dos dados banc√°rios para correla√ß√£o"""
    np.random.seed(42)
    
    # Dados banc√°rios simulados com sazonalidade
    datas = pd.date_range('2024-01-01', '2024-12-31', freq='D')
    
    # Simular impacto de fatores externos
    base_volume = 1000
    trend_anual = np.linspace(0, 200, len(datas))  # Crescimento anual
    sazonalidade_mensal = 100 * np.sin(2 * np.pi * np.arange(len(datas)) / 30)  # Ciclo mensal
    ruido = np.random.normal(0, 50, len(datas))
    
    volume_transacoes = base_volume + trend_anual + sazonalidade_mensal + ruido
    
    dados_bancarios = pd.DataFrame({
        'data': datas,
        'volume_transacoes': np.maximum(volume_transacoes, 100),  # Volume m√≠nimo
        'valor_medio_transacao': np.random.uniform(800, 1500, len(datas)),
        'taxa_aprovacao': np.random.uniform(88, 95, len(datas))
    })
    
    return dados_bancarios

def analisar_correlacao_externa(dados_externos, dados_bancarios):
    """Analisa correla√ß√£o entre dados externos e m√©tricas banc√°rias"""
    print("\nAnalisando correla√ß√£o com fatores externos...")
    
    # Merge dos dados por data
    dados_consolidados = dados_bancarios.merge(
        dados_externos['usd_brl'], on='data', how='left'
    )
    
    # Adicionar dados mensais (SELIC e IPCA)
    dados_consolidados['ano_mes'] = dados_consolidados['data'].dt.to_period('M')
    
    # Merge com SELIC
    selic_mensal = dados_externos['selic'].copy()
    selic_mensal['ano_mes'] = selic_mensal['data'].dt.to_period('M')
    dados_consolidados = dados_consolidados.merge(
        selic_mensal[['ano_mes', 'taxa_selic']], on='ano_mes', how='left'
    )
    
    # Merge com IPCA
    ipca_mensal = dados_externos['ipca'].copy()
    ipca_mensal['ano_mes'] = ipca_mensal['data'].dt.to_period('M')
    dados_consolidados = dados_consolidados.merge(
        ipca_mensal[['ano_mes', 'ipca_mensal']], on='ano_mes', how='left'
    )
    
    # Calcular correla√ß√µes
    correlacoes = dados_consolidados[['volume_transacoes', 'valor_medio_transacao', 
                                     'cotacao_usd', 'taxa_selic', 'ipca_mensal']].corr()
    
    print("Correla√ß√µes calculadas")
    
    return dados_consolidados, correlacoes

# ============================================================================
# DASHBOARD DE DADOS EXTERNOS
# ============================================================================

def gerar_dashboard_externos(dados_consolidados, correlacoes, dados_externos):
    """Dashboard com an√°lise de fatores externos"""
    print("\nGerando dashboard de dados externos...")
    
    # Configurar estilo do matplotlib
    plt.style.use('default')  # Usar estilo padr√£o para evitar problemas
    fig, axes = plt.subplots(3, 2, figsize=(16, 18))
    fig.suptitle('AN√ÅLISE DE FATORES EXTERNOS - BANVIC', fontsize=16, fontweight='bold', y=0.98)
    
    # 1. Evolu√ß√£o USD/BRL vs Volume de Transa√ß√µes
    ax1 = axes[0, 0]
    ax1_twin = ax1.twinx()
    
    line1 = ax1.plot(dados_consolidados['data'], dados_consolidados['volume_transacoes'], 
                     color='blue', label='Volume Transa√ß√µes', linewidth=1.5)
    line2 = ax1_twin.plot(dados_consolidados['data'], dados_consolidados['cotacao_usd'], 
                          color='red', label='USD/BRL', linewidth=1.5)
    
    ax1.set_title('Volume Transa√ß√µes vs Cota√ß√£o USD/BRL')
    ax1.set_ylabel('Volume Transa√ß√µes', color='blue')
    ax1_twin.set_ylabel('USD/BRL', color='red')
    ax1.grid(True, alpha=0.3)
    
    # Legenda combinada
    lines = line1 + line2
    labels = [l.get_label() for l in lines]
    ax1.legend(lines, labels, loc='upper left')
    
    # 2. Taxa SELIC ao longo do tempo
    ax2 = axes[0, 1]
    ax2.plot(dados_externos['selic']['data'], dados_externos['selic']['taxa_selic'], 
             marker='o', linewidth=2, color='green', markersize=4)
    ax2.set_title('Evolu√ß√£o da Taxa SELIC')
    ax2.set_ylabel('Taxa SELIC (%)')
    ax2.grid(True, alpha=0.3)
    ax2.tick_params(axis='x', rotation=45)
    
    # 3. IPCA Mensal e Acumulado
    ax3 = axes[1, 0]
    ax3_twin = ax3.twinx()
    
    bars = ax3.bar(dados_externos['ipca']['data'], dados_externos['ipca']['ipca_mensal'], 
                   alpha=0.7, color='orange', label='IPCA Mensal')
    line = ax3_twin.plot(dados_externos['ipca']['data'], dados_externos['ipca']['ipca_acumulado'], 
                         color='red', linewidth=2, marker='s', label='IPCA Acumulado')
    
    ax3.set_title('IPCA Mensal vs Acumulado')
    ax3.set_ylabel('IPCA Mensal (%)', color='orange')
    ax3_twin.set_ylabel('IPCA Acumulado (%)', color='red')
    ax3.tick_params(axis='x', rotation=45)
    ax3.grid(True, alpha=0.3)
    
    # Legenda
    ax3.legend(loc='upper left')
    ax3_twin.legend(loc='upper right')
    
    # 4. Mapa de Correla√ß√£o
    ax4 = axes[1, 1]
    correlacao_subset = correlacoes[['volume_transacoes', 'valor_medio_transacao']].loc[
        ['cotacao_usd', 'taxa_selic', 'ipca_mensal']
    ]
    
    sns.heatmap(correlacao_subset, annot=True, cmap='RdBu_r', center=0, 
                square=True, ax=ax4, cbar_kws={'shrink': 0.8})
    ax4.set_title('Correla√ß√£o: Fatores Externos vs M√©tricas Banc√°rias')
    ax4.set_xlabel('M√©tricas Banc√°rias')
    ax4.set_ylabel('Fatores Externos')
    
    # 5. An√°lise de Volatilidade USD/BRL
    ax5 = axes[2, 0]
    # Calcular volatilidade m√≥vel (janela de 30 dias)
    dados_consolidados['volatilidade_usd'] = dados_consolidados['cotacao_usd'].rolling(30).std()
    
    # Remover valores NaN da volatilidade
    dados_vol = dados_consolidados.dropna(subset=['volatilidade_usd'])
    
    ax5.fill_between(dados_vol['data'], dados_vol['volatilidade_usd'], 
                     alpha=0.6, color='purple')
    ax5.plot(dados_vol['data'], dados_vol['volatilidade_usd'], 
             color='indigo', linewidth=1)  # Mudan√ßa: 'darkpurple' -> 'indigo'
    ax5.set_title('Volatilidade USD/BRL (30 dias)')
    ax5.set_ylabel('Volatilidade')
    ax5.grid(True, alpha=0.3)
    ax5.tick_params(axis='x', rotation=45)
    
    # 6. Cen√°rios e Impactos
    ax6 = axes[2, 1]
    
    # Criar cen√°rios baseados nos dados
    cenarios_data = [
        'USD Baixo\n(<5.0)', 'USD M√©dio\n(5.0-5.5)', 'USD Alto\n(>5.5)',
        'SELIC Baixa\n(<11%)', 'SELIC Alta\n(>11%)'
    ]
    
    # Simular impactos nos volumes
    impactos_volume = [5, 0, -8, 10, -12]  # Varia√ß√£o percentual simulada
    
    colors = ['green' if x > 0 else 'red' if x < 0 else 'gray' for x in impactos_volume]
    bars = ax6.barh(cenarios_data, impactos_volume, color=colors, alpha=0.7)
    
    # Adicionar valores nas barras
    for i, (bar, valor) in enumerate(zip(bars, impactos_volume)):
        ax6.text(valor + (0.5 if valor >= 0 else -0.5), i, f'{valor:+.0f}%', 
                va='center', ha='left' if valor >= 0 else 'right', fontweight='bold')
    
    ax6.set_title('Impacto Estimado no Volume de Transa√ß√µes')
    ax6.set_xlabel('Varia√ß√£o no Volume (%)')
    ax6.axvline(x=0, color='black', linestyle='-', alpha=0.3)
    ax6.grid(True, alpha=0.3, axis='x')
    
    plt.tight_layout()
    plt.show()
    
    print("Dashboard de dados externos gerado")

# ============================================================================
# RELAT√ìRIO DE FATORES EXTERNOS
# ============================================================================

def gerar_relatorio_fatores_externos(dados_consolidados, correlacoes):
    """Relat√≥rio detalhado dos fatores externos"""
    print("\n" + "="*60)
    print("RELAT√ìRIO DE FATORES EXTERNOS")
    print("="*60)
    
    # Estat√≠sticas USD/BRL
    usd_stats = dados_consolidados['cotacao_usd'].describe()
    print(f"\nüí± AN√ÅLISE USD/BRL:")
    print(f"  ‚Ä¢ Cota√ß√£o M√©dia: R$ {usd_stats['mean']:.3f}")
    print(f"  ‚Ä¢ Varia√ß√£o (min-max): R$ {usd_stats['min']:.3f} - R$ {usd_stats['max']:.3f}")
    print(f"  ‚Ä¢ Volatilidade: {usd_stats['std']:.3f}")
    
    # Correla√ß√µes principais
    corr_volume_usd = correlacoes.loc['cotacao_usd', 'volume_transacoes']
    corr_valor_selic = correlacoes.loc['taxa_selic', 'valor_medio_transacao']
    
    print(f"\nüìä CORRELA√á√ïES PRINCIPAIS:")
    print(f"  ‚Ä¢ Volume vs USD/BRL: {corr_volume_usd:.3f}")
    print(f"  ‚Ä¢ Valor M√©dio vs SELIC: {corr_valor_selic:.3f}")
    
    # Classificar for√ßa das correla√ß√µes
    def classificar_correlacao(valor):
        abs_val = abs(valor)
        if abs_val > 0.7:
            return "FORTE"
        elif abs_val > 0.3:
            return "MODERADA"
        else:
            return "FRACA"
    
    print(f"  ‚Ä¢ For√ßa da correla√ß√£o Volume-USD: {classificar_correlacao(corr_volume_usd)}")
    print(f"  ‚Ä¢ For√ßa da correla√ß√£o Valor-SELIC: {classificar_correlacao(corr_valor_selic)}")
    
    # An√°lise de per√≠odos cr√≠ticos
    usd_alto = dados_consolidados[dados_consolidados['cotacao_usd'] > usd_stats['75%']]
    volume_medio_usd_alto = usd_alto['volume_transacoes'].mean()
    volume_geral = dados_consolidados['volume_transacoes'].mean()
    
    impacto_usd_alto = ((volume_medio_usd_alto - volume_geral) / volume_geral) * 100
    
    print(f"\nüéØ IMPACTOS IDENTIFICADOS:")
    print(f"  ‚Ä¢ Impacto USD Alto no Volume: {impacto_usd_alto:+.1f}%")
    
    # Recomenda√ß√µes
    print(f"\nüìã RECOMENDA√á√ïES:")
    
    if abs(corr_volume_usd) > 0.3:
        print(f"  ‚Ä¢ Monitorar c√¢mbio USD/BRL - correla√ß√£o {classificar_correlacao(corr_volume_usd).lower()}")
    
    if abs(corr_valor_selic) > 0.3:
        print(f"  ‚Ä¢ Acompanhar decis√µes COPOM - impacto na receita")
    
    print(f"  ‚Ä¢ Implementar alertas para volatilidade cambial > {usd_stats['std']*1.5:.3f}")
    print(f"  ‚Ä¢ Revisar estrat√©gias quando USD > R$ {usd_stats['75%']:.3f}")
    
    print(f"\n‚úÖ RELAT√ìRIO GERADO EM: {datetime.now().strftime('%d/%m/%Y √†s %H:%M:%S')}")

# ============================================================================
# EXECU√á√ÉO PRINCIPAL
# ============================================================================

def executar_analise_externa():
    """Pipeline principal de an√°lise externa"""
    print("üåê INICIANDO AN√ÅLISE DE DADOS EXTERNOS")
    
    # Buscar dados externos
    dados_externos = buscar_dados_externos()
    
    # Gerar dados banc√°rios sample
    dados_bancarios = gerar_dados_bancarios_sample()
    
    # An√°lise de correla√ß√£o
    dados_consolidados, correlacoes = analisar_correlacao_externa(dados_externos, dados_bancarios)
    
    # Dashboard
    gerar_dashboard_externos(dados_consolidados, correlacoes, dados_externos)
    
    # Relat√≥rio
    gerar_relatorio_fatores_externos(dados_consolidados, correlacoes)
    
    print("\nüéØ AN√ÅLISE DE DADOS EXTERNOS CONCLU√çDA!")
    
    return {
        'dados_externos': dados_externos,
        'dados_consolidados': dados_consolidados,
        'correlacoes': correlacoes
    }

# ============================================================================
# UTILIDADES PARA DADOS EXTERNOS
# ============================================================================

def salvar_dados_externos(dados_externos):
    """Salva dados externos localmente"""
    import os
    
    # Criar diret√≥rio se n√£o existir
    if not os.path.exists('./dados/externos'):
        os.makedirs('./dados/externos')
    
    # Salvar cada dataset
    dados_externos['usd_brl'].to_csv('./dados/externos/usd_brl.csv', index=False)
    dados_externos['selic'].to_csv('./dados/externos/selic.csv', index=False)
    dados_externos['ipca'].to_csv('./dados/externos/ipca.csv', index=False)
    
    print("Dados externos salvos localmente")

def carregar_dados_externos_salvos():
    """Carrega dados externos salvos localmente"""
    try:
        dados_externos = {
            'usd_brl': pd.read_csv('./dados/externos/usd_brl.csv'),
            'selic': pd.read_csv('./dados/externos/selic.csv'),
            'ipca': pd.read_csv('./dados/externos/ipca.csv')
        }
        
        # Converter datas
        for key in dados_externos:
            dados_externos[key]['data'] = pd.to_datetime(dados_externos[key]['data'])
        
        print("Dados externos carregados do cache local")
        return dados_externos
    
    except FileNotFoundError:
        print("Cache local n√£o encontrado, buscando dados novos...")
        return buscar_dados_externos()

# ============================================================================
# EXECU√á√ÉO
# ============================================================================

if __name__ == "__main__":
    resultado_externos = executar_analise_externa()

# Para ambiente Jupyter
try:
    if 'get_ipython' in globals():
        print("\nüéØ Executando no Jupyter...")
        resultado_externos = executar_analise_externa()
        print("‚úì Vari√°veis criadas: 'resultado_externos'")
except:
    pass
```
