```python
# KPIs E INDICADORES BANC√ÅRIOS - BANVIC
# Notebook para monitoramento de indicadores de performance
# Desenvolvido para acompanhamento gerencial

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta
import warnings
warnings.filterwarnings('ignore')

print("M√ìDULO DE KPIs E INDICADORES BANC√ÅRIOS")
print("="*50)

# ============================================================================
# CARREGAMENTO DOS DADOS
# ============================================================================

def carregar_dados():
    """Carrega dados base para c√°lculo dos KPIs"""
    print("\nCarregando dados para KPIs...")
    
    # Simula√ß√£o da base de dados real
    np.random.seed(42)
    n_trans = 12000
    data_base = datetime(2023, 1, 1)
    
    # Base de transa√ß√µes
    transacoes = pd.DataFrame({
        'cod_transacao': range(1, n_trans + 1),
        'data_transacao': pd.date_range(data_base, periods=n_trans, freq='1H'),
        'valor': np.random.lognormal(6.5, 1.5, n_trans),
        'tipo_transacao': np.random.choice([
            'PIX', 'TED', 'DOC', 'Transferencia', 'Deposito', 
            'Saque', 'Pagamento', 'Compra_Cartao'
        ], n_trans, p=[0.4, 0.1, 0.05, 0.15, 0.1, 0.08, 0.07, 0.05]),
        'canal': np.random.choice(['Mobile', 'Internet', 'Agencia', 'ATM'], 
                                n_trans, p=[0.5, 0.25, 0.15, 0.1]),
        'status': np.random.choice(['Aprovada', 'Rejeitada'], n_trans, p=[0.92, 0.08]),
        'num_conta': np.random.randint(1000, 1200, n_trans),
        'cod_agencia': np.random.randint(1, 21, n_trans)
    })
    
    # Base de contas
    contas = pd.DataFrame({
        'num_conta': range(1000, 1200),
        'saldo_atual': np.random.uniform(100, 50000, 200),
        'limite_credito': np.random.uniform(1000, 15000, 200),
        'tipo_conta': np.random.choice(['Corrente', 'Poupanca'], 200, p=[0.75, 0.25]),
        'data_abertura': pd.date_range('2020-01-01', '2023-01-01', periods=200),
        'status_conta': np.random.choice(['Ativa', 'Inativa'], 200, p=[0.95, 0.05]),
        'classificacao': np.random.choice(['A', 'B', 'C'], 200, p=[0.6, 0.3, 0.1])
    })
    
    # Base de ag√™ncias
    agencias = pd.DataFrame({
        'cod_agencia': range(1, 21),
        'nome_agencia': [f'Agencia_{str(i).zfill(3)}' for i in range(1, 21)],
        'cidade': np.random.choice(['S√£o Paulo', 'Rio de Janeiro', 'Belo Horizonte', 'Salvador'], 20),
        'funcionarios': np.random.randint(10, 50, 20),
        'meta_mensal': np.random.uniform(500000, 2000000, 20)
    })
    
    dados = {
        'transacoes': transacoes,
        'contas': contas,
        'agencias': agencias
    }
    
    print(f"‚úì {len(transacoes):,} transa√ß√µes carregadas")
    print(f"‚úì {len(contas)} contas ativas")
    print(f"‚úì {len(agencias)} ag√™ncias monitoradas")
    
    return dados

# ============================================================================
# C√ÅLCULO DOS KPIs PRINCIPAIS
# ============================================================================

def calcular_kpis(dados):
    """Calcula principais KPIs do banco"""
    print("\nCalculando KPIs principais...")
    
    trans = dados['transacoes']
    contas = dados['contas']
    agencias = dados['agencias']
    
    # Filtro apenas transa√ß√µes aprovadas
    trans_aprovadas = trans[trans['status'] == 'Aprovada'].copy()
    
    # KPIs de Volume
    kpis = {}
    
    # 1. Volume transacional
    kpis['volume_total'] = len(trans_aprovadas)
    kpis['valor_total'] = trans_aprovadas['valor'].sum()
    kpis['ticket_medio'] = trans_aprovadas['valor'].mean()
    kpis['mediana_ticket'] = trans_aprovadas['valor'].median()
    
    # 2. Taxa de aprova√ß√£o
    kpis['taxa_aprovacao'] = (len(trans_aprovadas) / len(trans)) * 100
    
    # 3. KPIs por per√≠odo
    trans_aprovadas['mes'] = trans_aprovadas['data_transacao'].dt.month
    kpis['volume_mensal_medio'] = trans_aprovadas.groupby('mes')['cod_transacao'].count().mean()
    kpis['valor_mensal_medio'] = trans_aprovadas.groupby('mes')['valor'].sum().mean()
    
    # 4. KPIs por canal
    kpis['canal_dominante'] = trans_aprovadas.groupby('canal')['cod_transacao'].count().idxmax()
    kpis['share_canal_dominante'] = (trans_aprovadas.groupby('canal')['cod_transacao'].count().max() / len(trans_aprovadas)) * 100
    
    # 5. KPIs de produto
    kpis['produto_lider'] = trans_aprovadas.groupby('tipo_transacao')['cod_transacao'].count().idxmax()
    kpis['concentracao_top3'] = (trans_aprovadas.groupby('tipo_transacao')['cod_transacao'].count()
                                .sort_values(ascending=False).head(3).sum() / len(trans_aprovadas)) * 100
    
    # 6. KPIs de contas
    contas_ativas = contas[contas['status_conta'] == 'Ativa']
    kpis['contas_ativas'] = len(contas_ativas)
    kpis['saldo_medio_contas'] = contas_ativas['saldo_atual'].mean()
    kpis['limite_medio_credito'] = contas_ativas['limite_credito'].mean()
    
    # 7. Performance por ag√™ncia
    performance_agencia = trans_aprovadas.groupby('cod_agencia').agg({
        'valor': 'sum',
        'cod_transacao': 'count'
    }).reset_index()
    
    kpis['agencia_top_volume'] = performance_agencia.loc[performance_agencia['cod_transacao'].idxmax(), 'cod_agencia']
    kpis['agencia_top_valor'] = performance_agencia.loc[performance_agencia['valor'].idxmax(), 'cod_agencia']
    
    return kpis, trans_aprovadas

def calcular_kpis_temporais(trans_aprovadas):
    """KPIs com evolu√ß√£o temporal"""
    print("Calculando KPIs temporais...")
    
    # Dados di√°rios
    kpis_diarios = trans_aprovadas.groupby(trans_aprovadas['data_transacao'].dt.date).agg({
        'cod_transacao': 'count',
        'valor': ['sum', 'mean']
    }).reset_index()
    
    kpis_diarios.columns = ['data', 'qtd_transacoes', 'valor_total', 'ticket_medio']
    kpis_diarios['data'] = pd.to_datetime(kpis_diarios['data'])
    
    # C√°lculo de tend√™ncias
    kpis_diarios = kpis_diarios.sort_values('data')
    kpis_diarios['volume_ma7'] = kpis_diarios['qtd_transacoes'].rolling(7).mean()
    kpis_diarios['valor_ma7'] = kpis_diarios['valor_total'].rolling(7).mean()
    
    # Taxa de crescimento semanal
    kpis_diarios['crescimento_volume'] = kpis_diarios['volume_ma7'].pct_change() * 100
    kpis_diarios['crescimento_valor'] = kpis_diarios['valor_ma7'].pct_change() * 100
    
    return kpis_diarios

# ============================================================================
# DASHBOARD DE KPIs
# ============================================================================

def gerar_dashboard_kpis(kpis, trans_aprovadas, kpis_temporais):
    """Dashboard executivo com principais KPIs"""
    print("\nGerando dashboard de KPIs...")
    
    plt.style.use('seaborn-v0_8')
    fig = plt.figure(figsize=(20, 16))
    
    # Layout do dashboard
    gs = fig.add_gridspec(4, 4, height_ratios=[1, 1.2, 1.2, 1], width_ratios=[1, 1, 1, 1])
    
    # T√≠tulo principal
    fig.suptitle('DASHBOARD DE KPIs BANC√ÅRIOS - BANVIC', fontsize=20, fontweight='bold', y=0.95)
    
    # ========== SE√á√ÉO 1: KPIs RESUMO (CARDS) ==========
    ax_cards = fig.add_subplot(gs[0, :])
    ax_cards.axis('off')
    
    # Cards de KPIs principais
    cards_data = [
        ("VOLUME TOTAL", f"{kpis['volume_total']:,}", "transa√ß√µes"),
        ("VALOR TOTAL", f"R$ {kpis['valor_total']/1000000:.1f}M", "movimentado"),
        ("TICKET M√âDIO", f"R$ {kpis['ticket_medio']:.0f}", "por transa√ß√£o"),
        ("TAXA APROVA√á√ÉO", f"{kpis['taxa_aprovacao']:.1f}%", "sucesso"),
        ("CANAL L√çDER", kpis['canal_dominante'], f"{kpis['share_canal_dominante']:.1f}% share")
    ]
    
    colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd']
    
    for i, (titulo, valor, subtitulo) in enumerate(cards_data):
        x_pos = i * 0.2 + 0.1
        # Fundo do card
        rect = plt.Rectangle((x_pos - 0.08, 0.2), 0.16, 0.6, 
                           facecolor=colors[i], alpha=0.1, transform=ax_cards.transAxes)
        ax_cards.add_patch(rect)
        
        # Textos
        ax_cards.text(x_pos, 0.7, titulo, fontsize=10, fontweight='bold', 
                     ha='center', transform=ax_cards.transAxes)
        ax_cards.text(x_pos, 0.5, valor, fontsize=14, fontweight='bold', 
                     color=colors[i], ha='center', transform=ax_cards.transAxes)
        ax_cards.text(x_pos, 0.3, subtitulo, fontsize=8, 
                     ha='center', transform=ax_cards.transAxes)
    
    # ========== SE√á√ÉO 2: GR√ÅFICOS PRINCIPAIS ==========
    
    # 1. Evolu√ß√£o temporal do volume
    ax1 = fig.add_subplot(gs[1, :2])
    ax1.plot(kpis_temporais['data'], kpis_temporais['qtd_transacoes'], 
             alpha=0.7, linewidth=1, label='Volume Di√°rio')
    ax1.plot(kpis_temporais['data'], kpis_temporais['volume_ma7'], 
             linewidth=2, color='red', label='M√©dia M√≥vel 7d')
    ax1.set_title('Evolu√ß√£o do Volume Transacional', fontsize=12, pad=15)
    ax1.set_ylabel('Qtd Transa√ß√µes')
    ax1.legend()
    ax1.grid(True, alpha=0.3)
    
    # 2. Evolu√ß√£o do valor
    ax2 = fig.add_subplot(gs[1, 2:])
    ax2.plot(kpis_temporais['data'], kpis_temporais['valor_total']/1000000, 
             alpha=0.7, linewidth=1, color='green', label='Valor Di√°rio')
    ax2.plot(kpis_temporais['data'], kpis_temporais['valor_ma7']/1000000, 
             linewidth=2, color='orange', label='M√©dia M√≥vel 7d')
    ax2.set_title('Evolu√ß√£o do Volume Financeiro', fontsize=12, pad=15)
    ax2.set_ylabel('R$ Milh√µes')
    ax2.legend()
    ax2.grid(True, alpha=0.3)
    
    # ========== SE√á√ÉO 3: AN√ÅLISES COMPLEMENTARES ==========
    
    # 3. Mix de produtos
    ax3 = fig.add_subplot(gs[2, 0])
    produtos = trans_aprovadas.groupby('tipo_transacao')['cod_transacao'].count().sort_values(ascending=False).head(6)
    wedges, texts, autotexts = ax3.pie(produtos.values, labels=produtos.index, 
                                      autopct='%1.1f%%', startangle=90)
    ax3.set_title('Mix de Produtos', fontsize=12, pad=15)
    for autotext in autotexts:
        autotext.set_fontsize(8)
    
    # 4. Performance por canal
    ax4 = fig.add_subplot(gs[2, 1])
    canais = trans_aprovadas.groupby('canal')['valor'].sum() / 1000000
    bars = ax4.bar(canais.index, canais.values, color=['skyblue', 'lightcoral', 'lightgreen', 'gold'])
    ax4.set_title('Volume por Canal (R$ MM)', fontsize=12, pad=15)
    ax4.set_ylabel('R$ Milh√µes')
    ax4.tick_params(axis='x', rotation=45)
    
    # Adicionar valores nas barras
    for bar in bars:
        height = bar.get_height()
        ax4.text(bar.get_x() + bar.get_width()/2., height + 0.1,
                f'R$ {height:.1f}M', ha='center', va='bottom', fontsize=9)
    
    # 5. Distribui√ß√£o de tickets
    ax5 = fig.add_subplot(gs[2, 2])
    ax5.hist(trans_aprovadas['valor'], bins=50, alpha=0.7, color='purple', edgecolor='black')
    ax5.axvline(kpis['ticket_medio'], color='red', linestyle='--', linewidth=2, label=f'M√©dia: R$ {kpis["ticket_medio"]:.0f}')
    ax5.axvline(kpis['mediana_ticket'], color='orange', linestyle='--', linewidth=2, label=f'Mediana: R$ {kpis["mediana_ticket"]:.0f}')
    ax5.set_title('Distribui√ß√£o dos Tickets', fontsize=12, pad=15)
    ax5.set_xlabel('Valor (R$)')
    ax5.set_ylabel('Frequ√™ncia')
    ax5.legend()
    ax5.set_xlim(0, trans_aprovadas['valor'].quantile(0.95))
    
    # 6. Top 5 ag√™ncias por volume
    ax6 = fig.add_subplot(gs[2, 3])
    top_agencias = trans_aprovadas.groupby('cod_agencia')['cod_transacao'].count().sort_values(ascending=False).head(5)
    bars = ax6.barh(range(len(top_agencias)), top_agencias.values, color='lightsteelblue')
    ax6.set_yticks(range(len(top_agencias)))
    ax6.set_yticklabels([f'AG_{ag:03d}' for ag in top_agencias.index])
    ax6.set_title('Top 5 Ag√™ncias (Volume)', fontsize=12, pad=15)
    ax6.set_xlabel('Qtd Transa√ß√µes')
    
    # Adicionar valores nas barras
    for i, bar in enumerate(bars):
        width = bar.get_width()
        ax6.text(width + 10, bar.get_y() + bar.get_height()/2,
                f'{int(width):,}', ha='left', va='center', fontsize=9)
    
    # ========== SE√á√ÉO 4: M√âTRICAS DE CRESCIMENTO ==========
    ax7 = fig.add_subplot(gs[3, :2])
    
    # Taxa de crescimento semanal
    crescimento_valido = kpis_temporais.dropna()
    ax7.plot(crescimento_valido['data'], crescimento_valido['crescimento_volume'], 
             label='Crescimento Volume', alpha=0.8, linewidth=1.5)
    ax7.axhline(y=0, color='black', linestyle='-', alpha=0.3)
    ax7.fill_between(crescimento_valido['data'], crescimento_valido['crescimento_volume'], 0, 
                    alpha=0.3, where=(crescimento_valido['crescimento_volume'] > 0), color='green')
    ax7.fill_between(crescimento_valido['data'], crescimento_valido['crescimento_volume'], 0, 
                    alpha=0.3, where=(crescimento_valido['crescimento_volume'] <= 0), color='red')
    ax7.set_title('Taxa de Crescimento Semanal (%)', fontsize=12, pad=15)
    ax7.set_ylabel('Crescimento (%)')
    ax7.legend()
    ax7.grid(True, alpha=0.3)
    
    # Tabela de m√©tricas resumo
    ax8 = fig.add_subplot(gs[3, 2:])
    ax8.axis('off')
    
    # Dados da tabela
    metricas_resumo = [
        ['M√©trica', 'Valor'],
        ['Contas Ativas', f"{kpis['contas_ativas']:,}"],
        ['Saldo M√©dio', f"R$ {kpis['saldo_medio_contas']:,.0f}"],
        ['Limite M√©dio', f"R$ {kpis['limite_medio_credito']:,.0f}"],
        ['Ag√™ncia Top Volume', f"AG_{kpis['agencia_top_volume']:03d}"],
        ['Ag√™ncia Top Valor', f"AG_{kpis['agencia_top_valor']:03d}"],
        ['Produto L√≠der', kpis['produto_lider']],
        ['Concentra√ß√£o Top 3', f"{kpis['concentracao_top3']:.1f}%"]
    ]
    
    # Criar tabela
    table = ax8.table(cellText=metricas_resumo[1:], colLabels=metricas_resumo[0],
                     cellLoc='left', loc='center', colWidths=[0.4, 0.6])
    table.auto_set_font_size(False)
    table.set_fontsize(10)
    table.scale(1.2, 1.5)
    
    # Estilizar cabe√ßalho
    for i in range(len(metricas_resumo[0])):
        table[(0, i)].set_facecolor('#4CAF50')
        table[(0, i)].set_text_props(weight='bold', color='white')
    
    # Estilizar c√©lulas
    for i in range(1, len(metricas_resumo)):
        for j in range(len(metricas_resumo[0])):
            if i % 2 == 0:
                table[(i, j)].set_facecolor('#f0f0f0')
    
    ax8.set_title('Resumo Executivo', fontsize=12, pad=20)
    
    plt.tight_layout()
    plt.show()
    
    print("‚úì Dashboard de KPIs gerado com sucesso")
    
    return kpis

# ============================================================================
# RELAT√ìRIO DETALHADO DOS KPIS
# ============================================================================

def gerar_relatorio_kpis(kpis, kpis_temporais):
    """Relat√≥rio textual detalhado dos KPIs"""
    print("\n" + "="*60)
    print("RELAT√ìRIO DETALHADO DE KPIs - BANVIC")
    print("="*60)
    
    print(f"\nüìä INDICADORES DE VOLUME:")
    print(f"  ‚Ä¢ Volume Total de Transa√ß√µes: {kpis['volume_total']:,}")
    print(f"  ‚Ä¢ Valor Total Movimentado: R$ {kpis['valor_total']:,.2f}")
    print(f"  ‚Ä¢ Ticket M√©dio: R$ {kpis['ticket_medio']:.2f}")
    print(f"  ‚Ä¢ Ticket Mediano: R$ {kpis['mediana_ticket']:.2f}")
    
    print(f"\nüìà INDICADORES DE PERFORMANCE:")
    print(f"  ‚Ä¢ Taxa de Aprova√ß√£o: {kpis['taxa_aprovacao']:.2f}%")
    print(f"  ‚Ä¢ Volume Mensal M√©dio: {kpis['volume_mensal_medio']:.0f} transa√ß√µes")
    print(f"  ‚Ä¢ Valor Mensal M√©dio: R$ {kpis['valor_mensal_medio']:,.2f}")
    
    print(f"\nüè¶ INDICADORES OPERACIONAIS:")
    print(f"  ‚Ä¢ Canal Dominante: {kpis['canal_dominante']} ({kpis['share_canal_dominante']:.1f}% share)")
    print(f"  ‚Ä¢ Produto L√≠der: {kpis['produto_lider']}")
    print(f"  ‚Ä¢ Concentra√ß√£o Top 3 Produtos: {kpis['concentracao_top3']:.1f}%")
    
    print(f"\nüí∞ INDICADORES DE CONTAS:")
    print(f"  ‚Ä¢ Contas Ativas: {kpis['contas_ativas']:,}")
    print(f"  ‚Ä¢ Saldo M√©dio por Conta: R$ {kpis['saldo_medio_contas']:,.2f}")
    print(f"  ‚Ä¢ Limite M√©dio de Cr√©dito: R$ {kpis['limite_medio_credito']:,.2f}")
    
    print(f"\nüè¢ INDICADORES DE AG√äNCIAS:")
    print(f"  ‚Ä¢ Ag√™ncia Top Volume: AG_{kpis['agencia_top_volume']:03d}")
    print(f"  ‚Ä¢ Ag√™ncia Top Valor: AG_{kpis['agencia_top_valor']:03d}")
    
    # An√°lise de tend√™ncia
    crescimento_recente = kpis_temporais['crescimento_volume'].tail(7).mean()
    if not np.isnan(crescimento_recente):
        print(f"\nüìä TEND√äNCIA RECENTE:")
        status_tendencia = "POSITIVA" if crescimento_recente > 0 else "NEGATIVA"
        print(f"  ‚Ä¢ Crescimento M√©dio (7 dias): {crescimento_recente:.2f}% - {status_tendencia}")
    
    # Alertas e recomenda√ß√µes
    print(f"\n‚ö†Ô∏è  ALERTAS E RECOMENDA√á√ïES:")
    
    if kpis['taxa_aprovacao'] < 90:
        print(f"  ‚Ä¢ Taxa de aprova√ß√£o baixa ({kpis['taxa_aprovacao']:.1f}%) - Investigar rejei√ß√µes")
    
    if kpis['concentracao_top3'] > 80:
        print(f"  ‚Ä¢ Alta concentra√ß√£o em poucos produtos ({kpis['concentracao_top3']:.1f}%) - Diversificar portf√≥lio")
    
    if kpis['share_canal_dominante'] > 60:
        print(f"  ‚Ä¢ Depend√™ncia excessiva do canal {kpis['canal_dominante']} - Equilibrar canais")
    
    print(f"\n‚úÖ RELAT√ìRIO GERADO EM: {datetime.now().strftime('%d/%m/%Y √†s %H:%M:%S')}")

# ============================================================================
# EXECU√á√ÉO PRINCIPAL
# ============================================================================

def executar_analise_kpis():
    """Pipeline principal de KPIs"""
    print("üöÄ INICIANDO AN√ÅLISE DE KPIs")
    
    # Carregar dados
    dados = carregar_dados()
    
    # Calcular KPIs
    kpis, trans_aprovadas = calcular_kpis(dados)
    kpis_temporais = calcular_kpis_temporais(trans_aprovadas)
    
    # Gerar dashboard
    gerar_dashboard_kpis(kpis, trans_aprovadas, kpis_temporais)
    
    # Relat√≥rio detalhado
    gerar_relatorio_kpis(kpis, kpis_temporais)
    
    print("\nüéØ AN√ÅLISE DE KPIs CONCLU√çDA COM SUCESSO!")
    
    return {
        'kpis': kpis,
        'dados': dados,
        'kpis_temporais': kpis_temporais,
        'trans_aprovadas': trans_aprovadas
    }

# Execu√ß√£o
if __name__ == "__main__":
    resultado_kpis = executar_analise_kpis()

# Para ambiente Jupyter
try:
    if 'get_ipython' in globals():
        print("\nüéØ Executando no Jupyter...")
        resultado_kpis = executar_analise_kpis()
        print("‚úì Vari√°veis criadas: 'resultado_kpis'")
except:
    pass
```
